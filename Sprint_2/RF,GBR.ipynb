# Catboost

```python
from catboost import CatBoostRegressor


model = CatBoostRegressor(
    iterations=3000,
    learning_rate=0.01,
    depth=10,
    loss_function='RMSE',
    l2_leaf_reg=5,
    cat_features=cat_features,
    verbose=500
)

# Train the model on your full training set and validate on the test set
model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test),
    plot=True  # Plot training process
)
```

```python
# catboost Predictions for the training set
y_train_pred = model.predict(X_train)

# Predictions for the train set
y_test_pred = model.predict(X_test)
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_train_pred)

# Calculate error metrics for the test set
mae_test = mean_absolute_error(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_test_pred)

print(f"Training Data Metrics:")
print(f"MAE: {mae_train}")
print(f"RMSE: {rmse_train}")
print(f"R: {r2_train}")

print(f"\nTest Data Metrics:")
print(f"MAE: {mae_test}")
print(f"RMSE: {rmse_test}")
print(f"R: {r2_test}")
```

# Gradient Boost Regressor

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


# Ensure all column names in X are strings
X.columns = X.columns.astype(str)

# Identify categorical features in X and encode them using LabelEncoder
categorical_cols = X.select_dtypes(include=['object']).columns
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical feature
for col in categorical_cols:
    X[col] = label_encoder.fit_transform(X[col])


X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)


gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)


gbr.fit(X_train, y_train)


y_train_pred = gbr.predict(X_train)


y_test_pred = gbr.predict(X_test)
```

```python
# Predictions for the train set
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_train_pred)

# Calculate error metrics for the test set
mae_test = mean_absolute_error(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_test_pred)

print(f"Training Data Metrics:")
print(f"MAE: {mae_train}")
print(f"RMSE: {rmse_train}")
print(f"R: {r2_train}")

print(f"\nTest Data Metrics:")
print(f"MAE: {mae_test}")
print(f"RMSE: {rmse_test}")
print(f"R: {r2_test}")
```
